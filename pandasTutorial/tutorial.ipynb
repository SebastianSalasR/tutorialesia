### Sorting data

```python
coffee.sort_values(by="Units Sold", ascending=False)
coffee.sort_values(by="Units Sold")
coffee.sort_values(by=["Units Sold", "Coffee Type"]) # If the amount of units sold it's the same for units sold, then it will sort by coffee type
```

### There are some cases where you want to iterate over rows using fow, but it's not recomemended becasuse you lose the memory eficiency of pandas

# Adding / Removing Columns

```python
coffee.head()
```

```python
coffee["price"] = 4.99
coffee.head()
```

```python
import numpy as np

coffee["new_price"] = np.where(coffee["Coffee Type"] == "Espresso", 3.99, 5.99) #If coffe type = espresso, price = 3.99, else price = 5.99
coffee.drop(columns=['price'], inplace=True) # With inplace, we modify the original dataframe
coffee
```

```python
#coffee_new = coffee
coffee_new = coffee.copy()
```

#### Note: This is weird because "coffee_new" is pointing to the same memory location as "coffee". Thats because if we change something in "coffee_new", it will also change in "coffee". To get rid of that problem, we need to creeate a new copy of the dataframe

```python
coffee_new["new_price"] = 4.99
coffee.head()
```

```python
coffee_new.head()
```

```python
coffee.head()
```

```python
coffee["revenue"] = coffee["Units Sold"]*coffee["new_price"]
coffee.head()
```

```python
coffee.rename(columns = {"new_price": "price"}, inplace=True)
coffee.head()
```

```python
bios["height_category"]  = bios["height_cm"].apply(lambda x: "short" if x < 165 else ("average" if x < 185 else "tall"))
bios
```

```python
def categorize_athlete(row):
    if row["height_cm"] < 175 and row["weight_kg"] < 70:
        return "Lightweight"
    elif row["height_cm"] < 185 or row["weight_kg"] <= 80:
        return "Middlewight"
    else:
        return "Heavyweight"
bios["Category"] = bios.apply(categorize_athlete, axis=1)
bios.head()
```
